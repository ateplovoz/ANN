## Выходные данные нейросети с одним слоем в пять нейронов

```python
import csv
import numpy as np
import matplotlib.pyplot as plt
```

```python
ann_data = np.array([])
with open('ann_data.csv') as datafile:
    reader = csv.reader(datafile)
    for row in reader:
        ann_data = np.append(ann_data, row)

ann_data = ann_data.astype(float)
ann_data = ann_data.reshape(20045, 4)
```

```python
src_data = np.array([])
with open('../data_ANN_full.csv') as datafile:
    reader = csv.reader(datafile)
    for row in reader:
        src_data = np.append(src_data, row)

src_data = src_data.astype(float)
src_data = src_data.reshape(23, 20045)

src_data_names = np.array([])
with open('../data_ANN_full_names.csv') as datafile:
    reader = csv.reader(datafile)
    for row in reader:
        src_data_names = np.append(src_data_names, row)

data_dict = dict(zip(src_data_names, src_data))
```

```python
f = plt.figure()
ax = plt.subplot()
ax.plot(data_dict['GT1'], data_dict['MainGenPower'], '.')
ax.plot(data_dict['GT1'], ann_data[:, 0]*1000, '.')
ax.grid()
plt.savefig('ann_data_gt1-mgp_raw.png', bbox_inches='tight')
plt.show()
```

```python
f = plt.figure()
ax = plt.subplot()
ax.plot(data_dict['GT1'], data_dict['TurbineExitTemp'], '.')
ax.plot(data_dict['GT1'], ann_data[:, 2], '.')
ax.grid()
plt.savefig('ann_data_gt1-tet_raw.png', bbox_inches='tight')
plt.show()
```

```python
f = plt.figure()
ax = plt.subplot()
ax.plot(data_dict['GT1'], data_dict['OutputPower'], '.')
ax.plot(data_dict['GT1'], ann_data[:, 1]*1000, '.')
ax.grid()
plt.savefig('ann_data_gt1-op_raw.png', bbox_inches='tight')
plt.show()
```

```python
f = plt.figure()
ax = plt.subplot()
ax.plot(data_dict['GT1'], data_dict['EngineSpeed'], '.')
ax.plot(data_dict['GT1'], ann_data[:, 3]*1000, '.')
ax.grid()
plt.savefig('ann_data_gt1-eng_raw.png', bbox_inches='tight')
plt.show()
```
